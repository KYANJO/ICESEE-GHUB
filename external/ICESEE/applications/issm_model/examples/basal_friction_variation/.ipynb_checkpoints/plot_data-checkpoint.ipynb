```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_file_path = 'data/new_data'
```

```python
file_path_true_nurged = f"{dataset_file_path}/_modelrun_datasets/true_nurged_states.h5"  
with h5py.File(file_path_true_nurged, 'r') as f:
    dset_nurged = f['nurged_state'][:]
    dset_true = f['true_state'][:]

file_path_obs = f"{dataset_file_path}/_modelrun_datasets/synthetic_obs.h5"
with h5py.File(file_path_obs, 'r') as f:
    dset_obs = f['hu_obs'][:] 

file_path_ens=f"{dataset_file_path}/_modelrun_datasets/icesee_ensemble_data.h5"
with h5py.File(file_path_ens, 'r') as f:
    # dset_ens = f['states'][:]
    dset_ens = f['ensemble'][:]

file_path_mean=f"{dataset_file_path}/_modelrun_datasets/icesee_enkf_mean.h5"
with h5py.File(file_path_mean, 'r') as f:
    dset_mean = f['mean'][:]
```

```python
nd,nens,nt = dset_ens.shape
print(f"nd: {nd}, nens: {nens}, nt: {nt}")

num_steps = nt-1
hdim = nd//3

# ensemble datasets
h_ens = dset_ens[0:hdim,:,:]
bed_ens = dset_ens[hdim:2*hdim,:,:]
friction_ens = dset_ens[2*hdim:3*hdim,:,:]

# true and nurged datasets
h_true= dset_true[0:hdim,:]
bed_true= dset_true[hdim:2*hdim,:]
friction_true= dset_true[2*hdim:3*hdim,:]

h_nurged= dset_nurged[0:hdim,:]
bed_nurged= dset_nurged[hdim:2*hdim,:]
friction_nurged= dset_nurged[2*hdim:3*hdim,:]

# mean datasets
h_mean= dset_mean[0:hdim,:]
bed_mean= dset_mean[hdim:2*hdim,:]
friction_mean= dset_mean[2*hdim:3*hdim,:]

# observations
h_obs = dset_obs[0:hdim,:]
bed_obs = dset_obs[hdim:2*hdim,:]
friction_obs = dset_obs[2*hdim:3*hdim,:]

h_idx = h_true.shape[1]//2
time = np.arange(num_steps+1)
```

```python
# plotting
htrue = h_true[h_idx,:-1]
hnurged = h_nurged[h_idx,:-1]

plt.figure(figsize=(12, 8))
# plt.plot(time, htrue, 'k-', label='True', linewidth=2)
plt.plot(time, hnurged, 'r--', label='Nurged', linewidth=2)
```
