```python
from pathlib import Path
from IPython.display import display, HTML, Image
import os

def find_repo_root():
    p = Path.cwd().resolve()
    while p != p.parent:
        if (p / "icesee_jupyter_book").exists():
            return p
        p = p.parent
    raise FileNotFoundError("Repo root not found.")

root = find_repo_root()
base = root / "icesee_jupyter_book" / "_static" / "generated"

# Find newest run across all examples
runs = []
if base.exists():
    for ex in base.iterdir():
        if ex.is_dir():
            for r in ex.iterdir():
                if r.is_dir():
                    runs.append(r)

if not runs:
    display(HTML("<h2> </h2><p>No runs found yet. Go to Run Center and click Run.</p>"))
else:
    runs.sort(key=lambda p: p.stat().st_mtime, reverse=True)
    run_dir = runs[0]
    display(HTML(f"<h2> </h2><p><b>Folder:</b> <code>{run_dir}</code></p>"))

    h5s = sorted((run_dir / "results").glob("*.h5"))
    pngs = sorted((run_dir / "figures").glob("*.png"))

    display(HTML(f"<p><b>Outputs:</b> {len(h5s)} H5, {len(pngs)} figures</p>"))
    if h5s:
        display(HTML("<ul>" + "".join([f"<li><code>{p.name}</code></li>" for p in h5s[:10]]) + "</ul>"))

    if (run_dir / "report.ipynb").exists():
        display(HTML("<p><b>Report:</b> <code>report.ipynb</code> (open from file browser)</p>"))

    if pngs:
        display(HTML("<hr><h3>Figures</h3>"))
        for p in pngs[:12]:
            display(Image(filename=str(p)))
    else:
        display(HTML("<p style='color:#777'>No figures found.</p>"))
```
